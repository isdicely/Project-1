# Anxiety and Depression in the US in the time of COVID 
This notebook gathers the baseline 2019 IAD data, start of Phase I IAD Household Pulse Survey data, and the most recent Household Pulse Survey data. The objective is to understand how mental health (as reflected in indicators of anxiety and depression) of the US population has been impacted by the COVID-19 pandemic.
These indicators are derived from responses to the first two questions of the eight-item Patient Health Questionnaire (PHQ-2) and the seven-item Generalized Anxiety Disorder (GAD-2) scale. 

### Sources:
* NHIS Early Release of Slected Mental Health Estimates: https://wwwn.cdc.gov/NHISDataQueryTool/ER_Quarterly/index_quarterly.html

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
# Read Baseline Comparison csv file
input_data_file = "mt_data/IAD_Baseline_Comparison.csv"
baseline_df = pd.read_csv(input_data_file)
```

```python
baseline_df.head(10)
```

```python
# plot this data on a line chart
plt.plot(baseline_df['Time Period'],baseline_df['Indicators of Depression and/or Anxiety'], marker='o')

# add labels and title
plt.xlabel('Date')
plt.ylabel('Respondents Reporting Indicators of Anxiety or Depression(%)')
plt.title('Anxiety and/or Depression Levels in the USA')
plt.show()
```

```python
# Generate a bar plot using PANDAS.

# Plot it, set a title & display the chart
baseline_df.plot(kind="bar", figsize=(12,4), rot=45)
plt.ylabel("Respondents Reporting Indicators of Anxiety or Depression(%)")
plt.title('Anxiety and/or Depression Levels in the USA')
plt.show()
```

```python
# Generate a bar plot using PYPLOT.
# Set up the x and y axis values
x_axis = np.arange(len(baseline_df))
y_axis = baseline_df['Indicators of Depression and/or Anxiety']
plt.figure(figsize=(12,6))
plt.bar(x_axis, y_axis, color ='xkcd:sea blue', align='center', width=0.5)

# Tell pyplot where we would like to place each of our x axis headers
headers = list(baseline_df['Time Period'])
plt.xticks(x_axis, headers, rotation=45)

# Give our chart some labels and a tile
plt.title("Anxiety/Depression Levels in the USA")
plt.xlabel("Time Period End Date")
plt.ylabel("Respondents Reporting Anxiety or Depression(%)")
plt.show()
```
