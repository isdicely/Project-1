```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import scipy.stats as st
```

```python
# Improt data

csv_path = "Indicators_of_Anxiety_or_Depression_Based_on_Reported_Frequency_of_Symptoms_During_Last_7_Days.csv"
data_file = pd.read_csv(csv_path)
```

```python
# Make dataframe
df = pd.DataFrame(data_file)
CA_df = df.loc[df["State"]== "California"]
CA_df
Dep_Dis = CA_df.loc[CA_df["Indicator"]=="Symptoms of Depressive Disorder"]
CA_df
Anx_Dis = CA_df.loc[CA_df["Indicator"]=="Symptoms of Anxiety Disorder"]
CA_df
Dep_Anx = CA_df.loc [CA_df["Indicator"] == "Symptoms of Anxiety Disorder or Depressive Disorder"]
```

```python
# labels
time_periods = Dep_Anx["Time Period"].to_list()
time_period_labels = Dep_Anx["Time Period Label"].to_list()

# Values
Dep_values = Dep_Dis["Value"].to_list()
Anx_values = Anx_Dis["Value"].to_list()
Dep_Anx_Values = Dep_Anx["Value"].to_list()

x = np.arange(len(time_periods)) # the label locations
width = 0.3  # the width of the bars
fig, ax = plt.subplots()

fig.set_figheight(8)
fig.set_figwidth(20)
depression = ax.bar(x - width/3, Dep_values, width, label="Depression")
anxiety = ax.bar(x + width/3, Anx_values, width, label="Anxiety")
depression_anxiety =  ax.bar(x + 3*width/3, Dep_Anx_Values, width, label="Depression & Anxiety")

ax.set_ylabel("Time Periods")
plt.xticks(rotation = 90)
ax.set_title("Symptomes of Mental Disorders Over Time")
ax.set_xticks(x)
ax.set_xticklabels(time_period_labels)
ax.legend()
```
