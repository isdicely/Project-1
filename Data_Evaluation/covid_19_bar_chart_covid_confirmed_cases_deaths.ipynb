```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.patches as mpatches
import os
import csv
```

```python
covidtotal = pd.read_csv("Resources/covid_19_data.csv") 
covidtotal
```

```python
covidedit = covidtotal
covidedit = covidedit.drop([0, 14, 33])
covidedit
```

```python
# 3-	Bar chart of COVID confirmed cases & deaths for 2020-current; Anne
# a.	All (US).
# b.	California, New York
```

```python
NYc = []
CAc = []
USc = []

NYd = []
CAd = []
USd = []
date = []

NYc = covidedit["Total NY Actual Cases"].tolist()
CAc = covidedit["CA Actual Cases"].tolist()
USc = covidedit["US Actual Cases"].tolist()

NYd = covidedit["Total NY Actual Deaths"].tolist()
CAd = covidedit["CA Actual Deaths"].tolist()
USd = covidedit["US Actual Deaths"].tolist()

date = covidedit["Date"].tolist()
```

```python
NYCAc = [
    NYc,
    CAc
]

plt.figure(figsize=(20,10))

X = np.arange(33)
plt.bar(X + 0.00, NYCAc[0], color = 'b', width = 0.4)
plt.bar(X + 0.25, NYCAc[1], color = 'g', width = 0.4)

plt.ylabel("Number of Confirmed  Cases")
plt.title("Total Number of Confirmed Cases in NY and CA")

plt.xticks(X, date, rotation ='vertical') 

blue_patch = mpatches.Patch(color='blue', label='NY')
green_patch = mpatches.Patch(color='green', label='CA')
plt.legend(handles=[blue_patch, green_patch])

plt.show()
```

```python
plt.figure(figsize=(20,10))

X = np.arange(33)
plt.bar(X + 0.00, USc, color = 'r', width = 0.4)

plt.ylabel("Number of Confirmed  Cases (Millions)")
plt.title("Total Number of Confirmed Cases in All US")

plt.xticks(X, date, rotation ='vertical') 

red_patch = mpatches.Patch(color='red', label='US')
plt.legend(handles=[red_patch])

plt.show()
```

```python
NYCAd = [
    NYd,
    CAd
]

plt.figure(figsize=(20,10))

X = np.arange(33)
plt.bar(X + 0.00, NYCAd[0], color = 'b', width = 0.4)
plt.bar(X + 0.25, NYCAd[1], color = 'g', width = 0.4)

plt.ylabel("Number of Deaths")
plt.title("Total Number of Deaths in NY and CA")

plt.xticks(X, date, rotation ='vertical') 

blue_patch = mpatches.Patch(color='blue', label='NY')
green_patch = mpatches.Patch(color='green', label='CA')
plt.legend(handles=[blue_patch, green_patch])

plt.show()
```
